```python
import csv
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import LinearSVC

def expected(codeID):
    code = ""
    if codeID == 0:
        code = "Stairs"
    elif codeID == 1:
        code = "Drive"
    elif codeID == 2:
        code = "Walk"
    elif codeID == 3:
        code = "Run"
    elif codeID == 4:
        code = "Sitting"
    return code
```

```python
logReg = LogisticRegression()
logReg.fit(trainAves, trainTarget)
predictionLog = logReg.predict(testAves)
i = 0
correctReg = 0
for i in range(len(testTarget)):
    if testTarget[i] != predictionLog[i]:
        print("Entry", i, "Expected:", expected(testTarget[i]), "Got:", expected(predictionLog[i]))
    else:
        correctReg += 1
        
print("Score:", correctReg, "out of:", len(testTarget))
```
